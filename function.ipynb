# **Functions and Variable Scope**

Scope defines where a variable can be accessed

Python has:

* **Local scope** – inside functions

* **Global Scope** – outside all functions

Helps avoid naming conflicts and bugs.

    x = "global" # Global variable

    def example():

        x = "local" # Local variable (only inside function)

        print(x)

    example()
    # Output: local

    print(x)
    # Output: global

Function parameters are local variables

    def greet(name):

        print(f"Hello, {name}!")

name is a local variable

Only exists inside the function

After the function ends, name is gone

Function return values belong to the scope where assigned to variables

    def get_length(text):

        return len(text)

    length = get_length("NLP") # 'length' is global

text is local to the function

length is global because we assigned the return value to it.

* Use parameters to pass data in

* Use return to send data out

* Avoid changing global variables inside functions
unless necessary

* Prefer modular functions with clear input/output

## 🔁 What is a **callback function**?

### ✅ A **callback function** is:
> A function that is **passed as an argument** to another function and is **called later**, usually after a task is completed.

A callback function is a function passed as an argument to another function.

It’s a way to **tell one function to call another** when something happens.

---

### 📦 Real-life analogy:
Imagine you're ordering pizza 🍕 and you say:

> “Call me when it’s ready.”

That **"call me"** part is the callback — you're giving the pizza shop your number (a function), and they’ll **call you back** when they’re done.

---

### 🔍 Basic Python Example:

```python
def greet(name):
    print(f"Hello, {name}!")

def process(callback):
    user = "Alice"
    callback(user)  # Call the callback function with user

process(greet)
```

### 🔄 What’s happening here?

- `greet()` is a function.
- `process()` takes a **function as a parameter**.
- `process(greet)` passes the `greet` function into `process`.
- Inside `process`, the `greet` function is **called back** later.

---

### 🛠 Real Use Case: Sorting with custom logic

```python
names = ["Charlie", "Alice", "Bob"]

# Sort by length of name using a callback
names.sort(key=lambda name: len(name))

print(names)  # ['Bob', 'Alice', 'Charlie']
```

Here, the **`lambda`** (or any function) is a **callback** that tells `sort()` how to compare values.

---

### 💡 When do we use callback functions?

- Event handling (in GUIs, web frameworks)
- Asynchronous programming
- Functional programming
- Custom logic (sorting, filtering, mapping)
- APIs and libraries that take user-defined behavior

---

> A **callback function** is a function you **pass to another function**, to be **executed later** — it gives your code **flexibility and reusability**.

**Example**

````python
def my_callback(result):
    print(f"The result is: {result}")

def my_function(data, callback):
    # Perform some operations on 'data'...
    processed_data = data * 2
    # Call the callback function with the processed data
    callback(processed_data)

my_function(5, my_callback)  # Output: The result is: 10
````

**How it Works**

1.my_callback: This is our callback function. It simply prints the result it receives.

2.my_function: Takes data and a callback as arguments.

3.Inside my_function: It processes the data and then calls the provided callback function, passing the processed data as an argument.

4.my_function(5, my_callback): We call my_function, passing 5 as data and my_callback as the callback function.
